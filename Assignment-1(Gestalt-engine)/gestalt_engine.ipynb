# Implementing Gestalt law of common fate
## The green lines depict that when many objects move in same direction, we perceive them to move together as a group

```python
import cv2
import numpy as np
from IPython.display import Image
# Read image
image = cv2.imread('gestalt.jpg')
 
# Convert image to grayscale
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
 
# Use canny edge detection to get the edges of the image to feed into the Hough lines function
edges = cv2.Canny(gray,50,180,apertureSize=3)
 
# Apply HoughLinesP method to get the coordinates of points lying on a line
lines_list =[]
lines = cv2.HoughLinesP(
            edges, # Input edge image
            1, # Distance resolution in pixels
            np.pi/180, # Angle resolution in radians
            threshold=100, 
            minLineLength=5, # Min allowed length of line
            maxLineGap=5 # Max allowed gap between line for joining them
            )
 
# Iterate over points
for points in lines:
    
    # Extracted points nested in the list
    x1,y1,x2,y2=points[0]
    
    # Draw the lines joing the points
    cv2.line(image,(x1,y1),(x2,y2),(0,255,0),2)
    # Append the points to a global list
    lines_list.append([(x1,y1),(x2,y2)])
     
# Save the result image
cv2.imwrite('detectedLines.png',image)
```
