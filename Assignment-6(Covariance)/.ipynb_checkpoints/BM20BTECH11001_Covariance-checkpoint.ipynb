# Calculating cross covariance with various lags from row 0 to row 499

```python
import cv2
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
image = cv2.imread('pic_1.jpeg')
image_2 = cv2.imread('pic_2.jpeg')
image = cv2.resize(image, (500, 500))
image_2 = cv2.resize(image_2, (500, 500))
```

```python
def shift_elements(arr, num, fill_value):
    result = np.empty_like(arr)
    if num > 0:
        result[:,:num] = fill_value
        result[:,num:] = arr[:,:-num]
    elif num < 0:
        result[:,num:] = fill_value
        result[:,:num] = arr[:,-num:]
    else:
        result[:,:] = arr
    return result
```

```python
b, g, r = cv2.split(image)
b1, g1, r1 = cv2.split(image_2)
```

```python
b_lag_1 = np.copy(b)
cov_means = []
for i in range(0, 500):
    shifted_elements = shift_elements(b_lag_1, i, 0)
    b_cov_lag_1 = np.cov(b, shifted_elements)
    cov_means.append(b_cov_lag_1.mean())
#cv2.imshow('Corr', np.concatenate((b_cov_lag_0, b_cov_lag_1), axis=1))
#cv2.waitKey(0)
#cv2.destroyAllWindows()
```

```python
plt.imshow(b);
```

```python
plt.plot(np.linspace(0, 499, 500), cov_means)
plt.xlabel('Lags')
plt.ylabel('Mean of covariances');
```

```python
plt.plot(np.linspace(0, 499, 500), cov_means);
plt.xlabel('Lag')
plt.ylabel('Mean of cross-covariance');
```
